```python
def label_salary(s):
    if '10,' in s or '25,' in s:
        return 'Low'
    elif '50,' in s or '75,' in s:
        return 'Medium'
    elif '100,' in s or '150,' in s:
        return 'High'
    else:
        return 'Very High'

df['salary_band'] = df['salary'].apply(label_salary)
```

**Converting Categorical Data to Numerical Data**

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
for column in df.columns:
    df[column] = LE.fit_transform(df[column])
```

```python
df.info()
```

```python
df.drop(['salary_band','post_date'],axis =1,inplace = True)
```

## Modeling

```python
final_res=[]
```

```python
x = df.drop('seniority_level',axis =1)
y = df.seniority_level
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train,y_train)
```

```python
y_pred_LR = LR.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Logistic Regression :')
print('Accuracy Score',accuracy_score(y_test,y_pred_LR))
acc_LR = accuracy_score(y_test,y_pred_LR)
final_res.append(acc_LR)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(x_train,y_train)
```

```python
y_pred_DT = DT.predict(x_test)
```

```python
print('Decision Tree :')
print('Accuracy Score ',accuracy_score(y_test,y_pred_DT))
acc_DT = accuracy_score(y_test,y_pred_DT)
final_res.append(acc_DT)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
RF.fit(x_train,y_train)
```

```python
y_pred_RF = RF.predict(x_test)
```

```python
print('Random Forest Classifier :')
print('Accuracy Score ',accuracy_score(y_test,y_pred_RF))
acc_RF = accuracy_score(y_test,y_pred_RF)
final_res.append(acc_RF)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
KNN.fit(x_train,y_train)
```

```python
y_pred_KNN = KNN.predict(x_test)
```

```python
print('K Nearest Neighbors ')
print('Accuracy Score ',accuracy_score(y_test,y_pred_KNN))
acc_KNN = accuracy_score(y_test,y_pred_KNN)
final_res.append(acc_KNN)
```

```python
from xgboost import XGBClassifier
XGB = XGBClassifier()
XGB.fit(x_train,y_train)
```

```python
y_pred_XGB = XGB.predict(x_test)
```

```python
print('XGBoost ')
print('Accuracy Score ',accuracy_score(y_test,y_pred_XGB))
acc_XGB = accuracy_score(y_test,y_pred_XGB)
final_res.append(acc_XGB)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GBC = GradientBoostingClassifier()
GBC.fit(x_train,y_train)
```

```python
y_pred_GBC = GBC.predict(x_test)
```

```python
print('Gradient Boosting ')
print('Accuracy Score ',accuracy_score(y_test,y_pred_GBC))
acc_GBC = accuracy_score(y_test,y_pred_GBC)
final_res.append(acc_GBC)
```

```python
from sklearn.svm import SVC
SV = SVC()
SV.fit(x_train,y_train)
```

```python
y_pred_SV = SV.predict(x_test)
```

```python
print('Support Vector Machine ')
print('Accuracy Score ',accuracy_score(y_test,y_pred_SV))
acc_SV = accuracy_score(y_test,y_pred_SV)
final_res.append(acc_SV)
```

## Comparing Algorithm

```python
plt.xticks(rotation = 90)
sns.barplot(x = ['Logistic Regression','Decision Tree', 'Random Forest','K-Nearest Neighbors','XGBoost','GradientBoosting','Support Vector Machine'],
            y = [accuracy_score(y_test,y_pred_LR),accuracy_score(y_test,y_pred_DT),accuracy_score(y_test,y_pred_RF),accuracy_score(y_test,y_pred_KNN),accuracy_score(y_test,y_pred_XGB),accuracy_score(y_test,y_pred_GBC),accuracy_score(y_test,y_pred_SV)]).set_title('Accuracy Score')
```

```python
final = np.array(final_res)
result = final.reshape(-1,1)
columns=['Accuracy']
index = ['Logistic Regression','Decision Tree', 'Random Forest','K-Nearest Neighbors','XGBoost','GradientBoosting','Support Vector Machine']
final_result = pd.DataFrame(result,index = index,columns = columns)
final_result.sort_values(by='Accuracy', ascending=False)
```
